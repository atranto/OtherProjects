```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_excel('TRsFullDataUM_V2.0.xlsx')
```

```python
dataset.head()
```

```python
cols = ['TR Code','TR Creation Date/Time','TR Source Location Code','TR Dest Location Code',
       'TR Source Location Postal Code','TR Dest Location Postal Code','OriginFull','DestinationFull',
        'TR Pickup - Event Day', 'PUDate','Distance','ClusterDistance']
data_1 = dataset.drop(cols, axis = 1)
```

```python
data_1.head()
```

```python
cols_2 = ['TR Gross Weight (KG)','Nb of Ship Units','TR Gross Volume (M3)']
data_2 = data_1.drop(cols_2, axis = 1)
data_2.head()
```

```python
data_2.shape[0]
```

```python
java_data = data_1.groupby(data_2.columns.tolist(), as_index = False).sum()
for i in java_data.loc[:, java_data.dtypes == 'object'].columns:
    java_data[i] = java_data[i].apply(lambda x: x.replace(',', '') if ',' in x else x)
java_data.head()
```

```python
java_data.to_csv('Aggregated data (for Q4-5).csv', index = False)
```
